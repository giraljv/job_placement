```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
color_target = ['#023047','#fb8500']
```

```python
df_processed = pd.read_csv('../data/processed/job_placement.csv')
df_raw = pd.read_csv('../data/raw/jb_data.csv')
```

```python
df_processed.select_dtypes(np.number).corr()['status'].sort_values(ascending=True).plot(kind='bar',color='#023047')
plt.ylabel('correlation_percentage')
plt.xlabel('features')
plt.title('variables con mayor impacto')
plt.savefig('./figures/vars_impact.png')
```

## Analisis de variables con mayor impacto y correlacion a la hora de obtener o no una seleccion

```python
im_vars = ['ssc_percentage','hsc_percentage','degree_percentage','work_experience','specialisation']
```

```python
sns.histplot(data=df_raw,x='ssc_percentage',hue='status',palette=color_target)
plt.title('SSC score result')
plt.savefig('./figures/ssc_percentage_hist.png')
```

### observaciones

- para pasar el SSC es importante tener un score de minimo 35.
- Vemos que las personas que suelen tener un bajo escore son más propensas a no tener el puesto.


```python
sns.histplot(data=df_raw,x='hsc_percentage',hue='status',palette=color_target)
plt.title('hsc score with placement status')
plt.savefig('./figures/hsc_score.png')
```

### observaciones
- HSC mark is the average of your aligned examination mark and your aligned school assessment mark, which is based on how your school went in the actual HSC exams and where you are ranked in your school.
- Vemos que también un bajo escore aquí tiene impacto a la hora de obtener un puesto.

```python
sns.histplot(data=df_raw,x='degree_percentage',hue='status',palette=color_target)
plt.title('degree score')
plt.show()
```

```python
sns.countplot(data = df_raw,x='work_experience',hue='status',palette=color_target)
plt.title('work_experience with placement status')
plt.savefig('./figures/work_experience.png')
```

### observaciones

- Tener experiencia laboral definitivamente te da mayor relevancia a la hora de obtener un puesto de trabajo y mantenerlo.
- Vemos que se contratan un gran número de personas sin experiencia, pero al momento de contratación, esta variable es muy importante por los reclutadores.

```python
sns.countplot(data = df_raw,x='specialisation',hue='status',palette=color_target)
plt.title('specialisation with status placement')
plt.show()
```

### Observaciones

- No hay tanta correlación en la specialisation como con los demás.
- Sin embargo, vemos que personas contratadas tienen una especialización de MKT y FIN .
- Las personas con una especialización de MKT y HR tuvieron misma oportunidad de ser contratados.
