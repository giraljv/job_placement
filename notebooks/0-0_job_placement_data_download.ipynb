# Caracteristicas

## variables 
- gender: male and female
- ssc_percentage: Senior secondary exams percentage 10th Grade
- ssc_board: Board of education for ssc exams
- hsc_percentage: Higher secondary exams percentage (12th Grade)    
- hsc_board: Board of education for hsc exams
- hsc_subject: Subject of study for hsc
- degree_percentage: Percentage of marks in undergrad degree
- undergrad_degree: Undergrad degree majors
- work_experience: Past work experience
- emp_test_percentage: Aptitude test percentage
- specialisation: Postgrad degree majors - (MBA specialization)
- mba_percent: Percentage of marks in MBA degree
- status: (TARGET) Status of placement. Placed / Not Placed

## Variables en español:
- género: masculino y femenino
- porcentaje de ssc: porcentaje de exámenes de educación secundaria superior (10º grado)
- consejo ssc: consejo de educación para exámenes de educación secundaria superior
- porcentaje de hsc: porcentaje de exámenes de educación secundaria superior (12º grado)
- consejo hsc: consejo de educación para exámenes de educación secundaria superior
- asignatura hsc: asignatura de estudio para educación secundaria superior
- porcentaje de grado: porcentaje de calificaciones en grado universitario
- título universitario: carrera universitaria
- experiencia laboral: experiencia laboral anterior
- porcentaje de prueba de empleo: porcentaje de prueba de aptitud
- especialización: carrera de posgrado - (especialización MBA)
- porcentaje de MBA: porcentaje de calificaciones en el grado de MBA
- estado: (OBJETIVO) Estado de colocación. Colocado / No colocado
